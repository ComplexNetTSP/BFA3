```python
import pandas as pd
import numpy as np
import quandl
quandl.ApiConfig.api_key = "y-zCjPnDPkRm_5tTzGRY"

import matplotlib.pylab
%matplotlib inline
```

```python
ticker_list = [
    'EURONEXT/EDEN',
    'EURONEXT/AC',
]
```

```python
df = pd.DataFrame()
for t in ticker_list:
    data = quandl.get(t, start_date='2017-01-01', end_date='2017-12-31')
    df[t+"_mean"] = data["Last"].rolling(20).mean()
    df[t+"_std"] = data["Last"].rolling(20).std()
    df[t+"_close"] = data["Last"]
    df[t+"_bup"] = df[t+"_mean"] + df[t+"_std"]
    df[t+"_bdown"] = df[t+"_mean"] - df[t+"_std"]
```

```python
df[['EURONEXT/EDEN_close','EURONEXT/EDEN_mean', 'EURONEXT/EDEN_bup', 'EURONEXT/EDEN_bdown']].plot()
```

```python
for t in ticker_list:
    for i in range(len(df)):
         if (df.iloc[i][t+"_bup"] != np.nan) and (df.iloc[i][t+"_bdown"] != np.nan):
            if df.iloc[i][t+"_close"] < df.iloc[i][t+"_bdown"]:
                print("Buy", t)
            if df.iloc[i][t+"_close"] > df.iloc[i][t+"_bup"]:
                print("Sell", t )
```
